```python
import h2o
import numpy as np

import pandas as pd

from h2o.estimators.deeplearning import H2OAutoEncoderEstimator
```

```python
h2o.init()
```

```python
def ngram(inp, window):
    """
    Function which windows the input matrix. Input matrix should be a numpy array.
    """
    
    rows = inp.shape[0]
    cols = inp.shape[1]
        
    res_rows = rows - window + 1
    
    newrows = res_rows
    newcols = window * cols
    
    res = np.empty(shape=(res_rows, newcols))
    for i in range(0, res_rows):
        if window - 1 > 0:

            new_row = np.concatenate([inp[j] for j in range(i, i+window)])
        else:
            new_row = inp[i]
        
        res[i] = new_row
        
    
    return res
```

```python
matrix = np.array(np.arange(1, 13)).reshape((4,3))
ngram(matrix, 4)
```
